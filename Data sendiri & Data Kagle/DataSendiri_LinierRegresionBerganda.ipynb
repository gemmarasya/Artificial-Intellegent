```
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```
data = {
     'pelanggan': [20,35,30,50,45,60,25,55,40,65,70,32,48,53,37],
    'jam_buka':  [6,8,7,9,8,10,6,9,7,10,11,7,8,9,6],
    'pendapatan':[200,350,300,480,450,600,250,520,390,650,700,310,470,530,360]
}

df = pd.DataFrame(data)
print(df)
```

```
# menentukan variabel
X = df[['pelanggan', 'jam_buka']]
y = df['pendapatan']
```

```
# split data train dan test
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.5, random_state=42)
```

```
# membuat model regresi
model = LinearRegression()
model.fit(X_train, y_train)
```

```
# melihat atau menghitung koefisien
print("Intercept (a):", model.intercept_)
print("Koefisien (b1,b2):", model.coef_)
# b1 → pengaruh pelanggan
# b2 → pengaruh jam buka
```

```
# prediksi data testing
y_pred = model.predict(X_test)

print("Prediksi:", y_pred)
print("Pendapatan Asli:", y_test.values)
```

```
# evaluasi model
import numpy as np
print("MAE:", metrics.mean_absolute_error(y_test, y_pred))
print("MSE:", metrics.mean_squared_error(y_test, y_pred))
print("RMSE:", np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```
