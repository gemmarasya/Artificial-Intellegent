```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```
data = {
    'langkah': [2000, 5000, 3000, 7000, 4000, 8000, 6000, 9000, 3500, 7500],
    'kalori':  [120, 300, 180, 420, 250, 480, 350, 520, 200, 450]
}

df = pd.DataFrame(data)
print(df)
```

```
x = df[['langkah']]
y = df['kalori']
```

```
x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42 )
```

```
model = LinearRegression()
model.fit(x_train, y_train)
```

```
print("Intercept (a):", model.intercept_)
print("Koefisien (b):", model.coef_[0])
```

```
y_pred = model.predict(x_test)

print("Data Testing:")
print(x_test)
print("Kalori Asli:", y_test.values)
print("Kalori Prediksi:", y_pred)
```

```
print("MAE:", metrics.mean_absolute_error(y_test, y_pred))
print("MSE:", metrics.mean_squared_error(y_test, y_pred))
print("RMSE:", np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

```
prediksi = model.predict([[12000]])
print("Prediksi kalori untuk 12000 langkah :", prediksi[0])
```
